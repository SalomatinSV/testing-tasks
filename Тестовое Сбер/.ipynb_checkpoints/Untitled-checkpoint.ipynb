# Предобработка и анализ данных

Задача бинарной классификации и состоит в том, чтобы для каждого action_id в action_test предсказать result как float от 0 до 1.

Файл person содержит всех уникальных людей (и соответствующие характеристики), которые выполняли действия с течением времени. Каждая строка в файле представляет уникального человека. У каждого человека есть уникальный person_id.

Файл action_train содержит все уникальные действия (и соответствующие характеристики действий), которые каждый человек выполнял в течение определенного времени. Каждая строка в файле действий представляет собой уникальное действие, выполненное человеком в определенный день. Каждое действие имеет уникальный идентификатор action_id. В файле содержится несколько различных категорий действий. Действия типа 1 отличаются от действий типа 2-7, поскольку известно больше характеристик, связанных с действиями типа 1 (всего девять), чем с действиями типа 2-7 (которые имеют только одну связанную характеристику).

```python
def percent_missing(df):
    missing_num = df[df.columns].isna().sum().sort_values(ascending=False)
    missing_perc = (df[df.columns].isna().sum()/len(df)*100).sort_values(ascending=False)
    missing_data = pd.concat([missing_num,missing_perc],keys=['Total','Percentage'],axis=1)
    missing_data = missing_data[missing_data['Percentage']>0]
       
    return missing_data
```

# Baseline

В качестве начальной модели возьмем решающее дерево и посмотрим, как работает данная модель с нашими данными.

```python
from sklearn import tree
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
X = group_action.drop(['result', 
                       'Unnamed: 0', 
                       'date',
                       'person_id'], # убираем именно person_id т.к. люди принадлежат group_1
                      axis = 1) 
y = group_action.result
```

Разобьем наши данные на тестовые и тренировочные.

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 777)
print(X.shape, X_train.shape, X_test.shape)
```

```python
clf = tree.DecisionTreeClassifier()
```

Подберем наилучшие параметры для решающего дерева

```python
parameters = {"criterion": ['gini', 'entropy'],
                      'max_depth' : range(1, 50, 5)}
```

```python
grid_search_cf_clf = GridSearchCV(clf, parameters, cv=3)
```

```python
grid_search_cf_clf.fit(X_train,y_train)
```

```python
grid_search_cf_clf.best_params_
```

У нас получилось довольно глубокое дерево и можно предположить, что оно переобучено.

```python
best_clf = grid_search_cf_clf.best_estimator_
```

```python
cross_val_score(best_clf, X_test, y_test, cv = 3)
```

```python
best_clf.score(X_test, y_test)
```

```python
y_pred = best_clf.predict(X_test)
```

```python
y_pred_prob = best_clf.predict_proba(X_test)
```

Рассмотрим как распределились предсказания на тестовой выборке:

```python
pd.Series(y_pred_prob[:,1]).hist()
```

Выбранная модель довольно чётко классифицировала наши данные. Примерно такое же соотношение результатов было дано изначально.

Рассмотрим, какие параметры оказались самыми важными для классификации.

```python
feature_importances = best_clf.feature_importances_
```

```python
feature_importances_df = pd.DataFrame({'features': list(X_train), 'feature_importances': feature_importances})
```

```python
feature_importances_df.sort_values('feature_importances', ascending = False)
```

Параметры group_1 и ts оказались наиболее важными,  как и было видно на графиках при анализе данных.

# Выбор метрик

Ввиду абстрактности нашей задачи, выбрить такие метрики как precision_score и recall_score не имеет смысла, поэтому основными метриками были выбраны AUC-ROC и log-loss. К тому же log-loss сильно штрафует переобученные модели, поэтому на него стоит обратить особое внимание.

```python
from sklearn.metrics import roc_curve, auc
sns.set(font_scale=1.5)
sns.set_color_codes("muted")
plt.figure(figsize=(10, 8))
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob[:,1], pos_label=1)
roc_auc= auc(fpr, tpr)
lw = 2
plt.plot(fpr, tpr, lw=lw, label='ROC curve (area = %0.2f)' % (roc_auc))
plt.plot([0, 1], [0, 1], lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('decisionTree ROC curve')
plt.legend(loc = "lower right")
plt.show()
```

Как мы видим, на графике ROC кривая близка к идеальной. Площадь покрытия составила 0.98

```python
from sklearn.metrics import log_loss
```

```python
dt_loss = log_loss(y_test, y_pred_prob[:,1])
round(dt_loss, 2)
```

 Среднее значение логарифмической функции потерь оказалась довольно высокой, что и следовало ожидать.

```python
result = pd.DataFrame({'log_loss': [round(dt_loss, 2)], 'roc_auc': [round(roc_auc, 2)]}, index = ['decisionTree'])
result
```

Однако, было замечено, что люди, относящиеся к той или иной группе, имели одинаковые характеристики, несмотря на тип действия. Возможно, если группы статичны и не зависят от конкретного типа действия, можно было бы заполнить пропущенные значения одинаковымми характеристиками, к которым предрасположена группа.
